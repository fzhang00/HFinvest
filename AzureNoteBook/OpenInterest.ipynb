# Comex Copper Option Open Interest

```sql
select DATE, Calls, Puts from 
(select [DATE], trim([Type]) as Type, [OpenInterest] from OI_VOL_CMX_CU_OPTION
where trim([TYPE]) in ('Calls', 'Puts') and [Name]='Copper Option') T
PIVOT 
(
      AVG(OpenInterest)
      for Type in (Calls, Puts)
) as pvt

SELECT [Date]
      ,[Contract]
      ,[CashBuyer]
      ,[CashSellerSettlement]
      ,[Month3_Buyer]
      ,[Month3_Seller]
      ,[Dec1_Buyer]
      ,[Dec1_Seller]
  FROM [Commodity_A1].[dbo].[LME_baseMetal_price] where contract='Copper'

select DATE, [Copper Financial Futures], [Copper Futures], [E-mini Copper Futures] from 
(select date, trim(Name) as Name, OpenInterest from OI_VOL_CMX_CU_FUTURE) T
PIVOT 
(
      AVG(OpenInterest)
      for Name in ([Copper Financial Futures], [Copper Futures], [E-mini Copper Futures])
) as pvt
```

# Comex Aluminum Future Open Interest

```sql
select DATE,  [Alumina FOB Australia (Platts) Futures],
                [Aluminium European Premium Duty-Paid (Metal Bulletin) Future],
                [Aluminium European Premium Duty-Unpaid (Metal Bulletin) Futu],
                [Aluminum A380 Alloy (Platts) Futures],
                [Aluminum Futures],
                [Aluminum Japan Premium (Platts) Futures],
                [Aluminum MW U.S. Transaction Premium Platts (25MT) Futures]
from 
(select date, Name, OpenInterest from OI_VOL_CMX_AL_FUTURE) T
PIVOT 
(
    AVG(OpenInterest)
    for Name in ([Alumina FOB Australia (Platts) Futures],
                [Aluminium European Premium Duty-Paid (Metal Bulletin) Future],
                [Aluminium European Premium Duty-Unpaid (Metal Bulletin) Futu],
                [Aluminum A380 Alloy (Platts) Futures],
                [Aluminum Futures],
                [Aluminum Japan Premium (Platts) Futures],
                [Aluminum MW U.S. Transaction Premium Platts (25MT) Futures])
) as pvt
```

# Comex Aluminum Option Open Interest

The only contract isAluminum MW U.S. Transaction Premium Platts (25MT) Average P

```sql
select DATE, Options, Calls, Puts from 
(select [DATE], Name, trim([Type]) as Type, [OpenInterest] from OI_VOL_CMX_AL_OPTION
where trim([TYPE]) in ('Options', 'Calls', 'Puts')) T
PIVOT 
(
      AVG(OpenInterest)
      for Type in (Options, Calls, Puts)
) as pvt
```

# Comex Gold Future Open Interest

```sql
select date, [E-mini Gold Futures],
            [Gold Futures],
            [Micro Gold Futures],
            [Shanghai Gold (CNH) Futures],
            [Shanghai Gold (USD) Futures]
from 
    (select Date, trim(Name) as Name, OpenInterest from OI_VOL_CMX_AU_FUTURE) p
pivot(
    AVG(OpenInterest)
    for name in ([E-mini Gold Futures],
            [Gold Futures],
            [Micro Gold Futures],
            [Shanghai Gold (CNH) Futures],
            [Shanghai Gold (USD) Futures])
) pvt
```

# Comex Gold Option Open Interest

Available Types are: Gold option, 1 week, 2 weeks, ... 5 weeks

Selected data chose Gold option only. 

```sql
select DATE, Calls, Puts from 
(
    select [DATE], trim(Name) as Name,  trim([Type]) as Type, [OpenInterest] from OI_VOL_CMX_AU_OPTION where Name like 'Gold Option'
) T
PIVOT 
(
      AVG(OpenInterest)
      for Type in (Calls, Puts)
) as pvt
```
